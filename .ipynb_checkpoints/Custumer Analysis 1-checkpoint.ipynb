```python
import pandas as pd 
import numpy as np
```

```python
df1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
df2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
df3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
df1.head()
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
cols1 = []
for column in df1.columns:
    cols1.append(column.lower())
df1.columns = cols1

cols2 = []
for column in df2.columns:
    cols2.append(column.lower())
df2.columns = cols2

cols3 = []
for column in df3.columns:
    cols3.append(column.lower())
df3.columns = cols3

df3 = df3.rename(columns={'state':'st'})
```

```python
df2 = df2[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

df3 = df3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
```

```python
list1 = df1['customer lifetime value']
listdf1 = []
test = []

for value in list1:
    listdf1.append(str((value)))

for value in listdf1:
    test.append(value.replace("%", ""))

test = pd.to_numeric(test, errors='coerce')

df1['customer lifetime value'] = test/100

list2 = df2['customer lifetime value']
listdf2 = []
test2 = []

for value in list2:
    listdf2.append(str((value)))

for value in listdf2:
    test2.append(value.replace("%", ""))

test2 = pd.to_numeric(test2, errors='coerce')

df2['customer lifetime value'] = test2/100
```

```python
df = pd.concat([df1, df2, df3], axis = 0)

df.shape
```
